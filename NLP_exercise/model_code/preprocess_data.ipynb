```python
import pandas as pd
import os
directory_path = os.getcwd() + "\\data"
os.path.exists(directory_path)
```

```python
with open(directory_path + "\\rawData\\unlabeledTrainData.tsv", "r", encoding="utf-8") as f:
    unlabeledTrain = [line.strip().split("\t") for line in f.readlines() if len(line.strip().split("\t")) == 2]

with open(directory_path + "\\rawData\\labeledTrainData.tsv", "r", encoding="utf-8") as f:
    labeledTrain = [line.strip().split("\t") for line in f.readlines() if len(line.strip().split("\t")) == 3]
```

```python
unlabel_data = pd.DataFrame(unlabeledTrain[1: ], columns=unlabeledTrain[0])
label_data = pd.DataFrame(labeledTrain[1: ], columns=labeledTrain[0])
```

```python
unlabel_data.head()
```

```python
label_data.head()
```

```python
def getRate(subject):
    split_list = subject[1:-1].split("_")
    return int(split_list[1])

label_data["rate"] = label_data["id"].apply(getRate)
label_data.head()
```

```python
from bs4 import BeautifulSoup

def cleanReview(subject):
    beau_soup = BeautifulSoup(subject)
    new_subject = beau_soup.get_text()
    new_subject = new_subject.strip().split(" ")
    new_subject = [word.lower() for word in new_subject]
    new_subject = " ".join(new_subject)
    return new_subject

unlabel_data["review"] = unlabel_data["review"].apply(cleanReview)
label_data["review"] = label_data["review"].apply(cleanReview)
```
